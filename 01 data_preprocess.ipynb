```python
def get_info(info, books, is_2017=True):
    num = []
    dif = []
    sold = []
    genre = []
    price = []
    shop = []
    word_list = []
    # books是info中2017/18年的所有信息
    for book in books:
        # 将info中的信息，通过book ID 进行连接，插入新列&信息
        # book 是包含某用户读过的所有书ID的数组, 所以word_list是某用户当年读的所有书包含的单词的集合
        # get data from table "words"
        word_list_1 = words.loc[info['Book ID'] == book, 'Words in Book']
        word_list_1 = word_list_1.values[0].split('|')
        word_list += word_list_1
        # word_list.append(word_list_1)
        
        # get data from table "info"
        n1 = info.loc[info['Book ID'] == book, 'Number Of Words']
        d1 = info.loc[info['Book ID'] == book, 'Difficulty (Reader suggested)']
        s1 = info.loc[info['Book ID'] == book, 'Number Sold']
        g1 = info.loc[info['Book ID'] == book, 'Book Genre']
        shop1 = info.loc[info['Book ID'] == book, 'Most Sold At']
        if is_2017:
            p1 = info.loc[info['Book ID'] == book, 'Price (2017)']
        else:
            p1 = info.loc[info['Book ID'] == book, 'Price (2018)']
        # 将字数以500字的倍数进行归类
        n1 = n1.values[0] // 500
        # 放到一个新的list里，一会连续变量离散化
        num.append(n1)
        # 将销量以500本的倍数进行归类
        s1 = s1.values[0] // 500
        # 放到一个新的list里，一会连续变量离散化
        sold.append(s1)
        # 因为2017和2018的价格是分开写的，同一本书在17和18年的价格不一样
        # 为了使价格在one-hot之后具有可比性，同一本书的价格还能一样/差不多（同一个维度上one-hot的位置一样）
        # 就要统一 两年价格以“5块的倍数进行归类”后的维度
        if p1.values[0] >=49:
            p1 = 49
        else:
            p1 = p1.values[0]
        p1 = p1 // 5
        price.append(p1)
        # 取出，一会插入data表
        dif.append(d1.values[0])
        genre.append(g1.values[0])
        shop.append(shop1.values[0])
    # 每一个data(user)表中，有很多books，上边的for循环得到了其中包含的每一个book的信息list，
    # 循环得到所有的信息，比如difficulty，number sold等等，所以循环之外要把这些list中的数据合成一个datafram中的一个数据
    dif = ', '.join(str(x) for x in dif)
    num = ', '.join(str(x) for x in num)
    sold = ', '.join(str(x) for x in sold)
    genre = ', '.join(str(x) for x in genre)
    price = ', '.join(str(x) for x in price)
    shop = ', '.join(str(x) for x in shop)
    return dif, num, sold, genre, price, shop, word_list
```

```python
for index, row in data.iterrows():
    books_2017 = row['User Read Books (2017)']
    books_2017 = [int(s) for s in books_2017.split(', ')]
    books_2018 =  row['User Read Books (2018)']
    books_2018 = [int(s) for s in books_2018.split(', ')]
    
    dif, num, sold, genre, price, shop, word_list = get_info(info, books_2017)
    # 将words,info表的信息插入data表中
    data.at[index, 'User Read Words (2017)'] = num
    data.at[index, 'User Read Difficulty (2017)'] = dif
    data.at[index, 'User Read sales (2017)'] = sold
    data.at[index, 'User Read genre (2017)'] = genre
    data.at[index, 'User Read price (2017)'] = price
    data.at[index, 'User Read shop (2017)'] = shop
    # 新建一个feature， 一年读了几本书
    data.at[index, 'User Read number (2017)'] = str(len(books_2017))
    data.at[index, 'book contain words (2017)'] = word_list

    dif2018, num2018, sold2018, genre2018, price2018, shop2018, word_list2018 = get_info(info, books_2018, is_2017=False)
    data.at[index, 'User Read Words (2018)'] = num2018
    data.at[index, 'User Read Difficulty (2018)'] = dif2018
    data.at[index, 'User Read sales (2018)'] = sold2018
    data.at[index, 'User Read genre (2018)'] = genre2018
    data.at[index, 'User Read price (2018)'] = price2018
    data.at[index, 'User Read shop (2018)'] = shop2018
    data.at[index, 'User Read number (2018)'] = str(len(books_2018))
    data.at[index, 'book contain words (2018)'] = word_list2018
```
