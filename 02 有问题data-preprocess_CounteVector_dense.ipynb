**CounterVectorizer：记录每行中每个单词在所有单词中出现的次数**。会得到很稀疏的矩阵，10000*所有单词的个数（比35488维要大），而大多数位置都是0.

** It takes a loooong time...**

https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.CountVectorizer.html

- lowercase : boolean, True by default. 这里全是数字，所以要给改成False。
- tokenizer : callable or None (default). Override the string tokenization step while preserving the preprocessing and n-grams generation steps. Only applies if analyzer == 'word'. 不是word是数字，所以用匿名函数lamda映射成本身

#### 只剩下15个feature，不太对啊。。

CountVector会把str中的每个词分出来，而如果str中如项目中，文字是用数字来代替的话，CountVector就会把一个一个的数字切出来，而不是把一串数字看作一个整体，就导致了只有数字加符号的15维这种问题。所以，显然，数字是不能用CountVector的。


- Transform a **count matrix** to a normalized tf or tf-idf representation
- smooth_idf : boolean, default=True。Smooth idf weights by adding one to document frequencies, as if an extra document was seen containing every term in the collection exactly once. **Prevents zero divisions**. 因为这里只有一个文章，所以设为False。

```python
transformer = TfidfTransformer(smooth_idf=False)
book_words_2017_tf_idf = transformer.fit_transform(book_words_2017_count_dense)
book_words_2017_tf_idf = book_words_2017_tf_idf.toarray()
```

- **Non-Negative Matrix Factorization (NMF)**
- Find two non-negative matrices (W, H) whose product approximates the non- negative matrix X. This factorization can be used for example for **dimensionality reduction, source separation or topic extraction**.
- ‘random’: non-negative random matrices, scaled with: **sqrt(X.mean() / n_components)**
- https://blog.csdn.net/ACdreamers/article/details/44663421
